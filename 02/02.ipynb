```python
file = open('input/input.txt', 'r')
lines = file.readlines()
```

```python
#a:
safe_lines = 0
for line in lines:
    asc = False
    split = line.split()
    if (int(split[0]) - int(split[-1])) > 0:
        asc = False
    elif (int(split[0]) - int(split[-1])) < 0:
        asc = True
    else:
        continue

    for i, x in enumerate(split):
        if i == len(split)-1:
            safe_lines += 1
            break
        diff = int(x) - int(split[i+1])
        if abs(diff) > 3 or (asc and diff >= 0) or (not asc and diff <= 0):
            break
print(safe_lines)
```

```python
#b:
safe_lines = 0
for k, line in enumerate(lines):
    split_orig = line.split()
    safe = False

    # remove one item and check (Brute Force :()
    for j, y in enumerate(split_orig):
        split = split_orig.copy()
        split.pop(j)

        asc = False
        if (int(split[0]) - int(split[-1])) > 0:
            asc = False
        elif (int(split[0]) - int(split[-1])) < 0:
            asc = True
        else:
            continue

        for i, x in enumerate(split):
            if i == len(split)-1:
                safe = True
                break
            diff = int(x) - int(split[i+1])
            if abs(diff) > 3 or (asc and diff >= 0) or (not asc and diff <= 0):
                break

        if safe:
            print(str(k+1)+" - "+str(split)+ ": Safe")
            safe_lines += 1
            break

    if not safe:
        print(str(k+1)+" - "+str(split_orig)+ ": Unsafe")
print("Safe lines: "+ str(safe_lines))
```
