### Part B:

```python
def check_possible_operations(orig_result, result, operands, right_side=-1):
    if len(operands) == 1 and result == int(operands[0]):
        return True
    if len(str(result)) > sum(len(s) for s in operands):
        # result is bigger than multiplication of all operands
        return False
    else:
        # remove last operand
        remaining_operands = operands.copy()
        last_operand = int(remaining_operands.pop())
        if right_side == -1:
            new_right_side = last_operand
        # check mult
        mult = False
        if result % last_operand == 0:
            if right_side != -1:
                new_right_side = last_operand*right_side
            mult = check_possible_operations(orig_result, result // last_operand, remaining_operands, new_right_side)
        # check add
        if right_side != -1:
            new_right_side = last_operand+right_side
        add = check_possible_operations(orig_result, result - last_operand, remaining_operands, new_right_side)
        # check concatenation
        conc = False
        if str(orig_result).endswith(str(right_side)):
            if right_side != -1:
                new_right_side = int(str(last_operand)+str(right_side))
            print(orig_result, int(str(orig_result)[: -len(str(right_side))]), remaining_operands, right_side, new_right_side )
            conc = check_possible_operations(orig_result, int(str(orig_result)[: -len(str(right_side))]), operands, new_right_side)
        return (mult or add or conc)
```

```python
#read input:
calibration_result = 0
for line in lines:
    result, operands = line.strip().split(': ')
    result = int(result)
    operands = operands.split()

    if check_possible_operations(result, result, operands):
        calibration_result += result

print(calibration_result)

```
