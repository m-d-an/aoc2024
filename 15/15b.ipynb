```python
file = open('input/input.txt', 'r')
lines = file.readlines()
```

```python
max_y = len(lines)
max_x = len(lines[0].strip())*2
```

```python
walls = []
boxes = []
robot = None
movements = []
```

```python
def print_map():
    for y in range(max_y):
        line = ""
        for x in range(max_x):
            if (x,y) in walls:
                line += '#'
            elif ((x,y),'l') in boxes:
                line += '['
            elif ((x,y),'r') in boxes:
                line += ']'
            elif (x,y) == robot:
                line += '@'
            else:
                line += '.'
        print(line)
```

```python

for y,line in enumerate(lines):
    if line == "\n":
        max_y = y
        continue
    for x, c in enumerate(line.strip()):
        if c == '#':
            walls.append((2*x,y))
            walls.append((2*x+1,y))
        elif c == 'O':
            boxes.append(((2*x,y),'l'))
            boxes.append(((2*x+1,y),'r'))
        elif c == '@':
            robot = (2*x,y)
        elif c == '<':
            movements.append((-1,0))
        elif c == '^':
            movements.append((0,-1))            
        elif c == '>':
            movements.append((1,0))
        elif c == 'v':
            movements.append((0,1))
```

```python
box_movements = []
```

```python
def double_move(new_pos_l, new_pos_r, move):
    boxes.remove(((new_pos_l[0]-move[0],new_pos_l[1]-move[1]),'l'))
    boxes.append((new_pos_l, 'l'))
    boxes.remove(((new_pos_r[0]-move[0],new_pos_r[1]-move[1]),'r'))
    boxes.append((new_pos_r, 'r'))
```

```python
def is_blocked(pos, move, side):
    if pos in walls:
        return True
    if side == 'l' and move == (1,0):
        return False
    if side == 'r' and move == (-1,0):
        return False
    if (pos,'l') in boxes:
        new_pos_l = (pos[0]+move[0],pos[1]+move[1])
        new_pos_r = (pos[0]+move[0]+1,pos[1]+move[1])
    elif (pos,'r') in boxes:
        new_pos_l = (pos[0]+move[0]-1,pos[1]+move[1])
        new_pos_r = (pos[0]+move[0],pos[1]+move[1])
    else:
        return False   
    
    if is_blocked(new_pos_l, move, 'l') or is_blocked(new_pos_r, move, 'r'):
        return True
    else:
        box_movements.append((new_pos_l,new_pos_r,move))
        return False
```

```python
for move in movements:

    # new pos after roboter movement
    new_pos = (robot[0]+move[0], robot[1]+move[1])

    # check if there is a wall
    if new_pos in walls:
        continue

    # check if there is a box
    if (new_pos,'l') in boxes:
        new_pos_l = (new_pos[0]+move[0],new_pos[1]+move[1])
        new_pos_r = (new_pos[0]+move[0]+1,new_pos[1]+move[1])
    elif (new_pos,'r') in boxes:
        new_pos_l = (new_pos[0]+move[0]-1,new_pos[1]+move[1])
        new_pos_r = (new_pos[0]+move[0],new_pos[1]+move[1])
    else:
        # if there is no box, just move
        robot = new_pos
        continue
    # check if box can be pushed or way is blocked
    box_movements = []
    if is_blocked(new_pos_l, move, 'l') or is_blocked(new_pos_r, move, 'r'):
        continue
    else:
        box_movements.append((new_pos_l,new_pos_r,move))
        for bm in set(box_movements):
             double_move(bm[0], bm[1], bm[2])
        robot = new_pos

print_map()
```

```python
# compute GPS coordinates
gps_sum = 0
for box in boxes:
    if box[1] == 'l':
        gps_sum += 100 * box[0][1] + box[0][0]
print(gps_sum)
```
